```python
flipkart_html = bs(flipkartPage, "html.parser") 
#bs is alias here for beautiful Soup which is responsible to beautify html tags so that it is little bit readable
```

```python
#we got a little bit structured format
flipkart_html
```

```python
#findall function in beutiful soup will help us to find something by which we are going to get the url
#here class id _1AtVbe is a id to reach out to any of the product
#beautiful soup will help us to find a division and inside division it will try to find for class
flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"}) 
```

```python
#As it returned a list we can check how many the number of iphone  acategory in the list
len(flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"}))
```

```python
bigboxes=flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"})
```

```python
#we can loop through the bigboxes(which is different iphones categories)
for i in bigboxes:
    print(i)
```

```python
#deleted first 3 boxes
del bigboxes[0:3]
```

```python
#if we want information about the very first box(means iphone category)
box=bigboxes[0]
```

```python
#we need to go to a particular box
#Now next task is to find href(what is the url for that box)
#here href is the key and has a value pair
#with the help of this url we will be able to reach inside that particular category of iphone with click
#we will apppend this url with the base url of flipkart
box.div.div.div.a['href']
```

```python
#we will apppend this url with the base url of flipkart
#This is the url with which we can reach inside particular category of iphone which is a product link
productLink="https://www.flipkart.com" + box.div.div.div.a['href']
```

```python
productLink
```

```python
#with the help of this product link we can go reach to particular category page of iphone
#Now our target is on that page reach out to scrap the reviews  which is again in the form of boxes 
prodRes = requests.get(productLink) #request to pass the product link
prodRes
```
